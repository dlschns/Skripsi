```python
def load_model(path):
    # Recreate the exact same model, including its weights and the optimizer
    load_model = tf.keras.models.load_model(path)

    # Show the model architecture
    load_model.summary()
    
    return load_model
```

## Print Predictions

```python
def Pengujian(df, model_path):
    model = load_model(model_path)
    features = np.array(df.columns.difference(['code', 'label citra']))
    
    test_ds = (
                tf.data.Dataset.from_tensors(
                    (
                        tf.cast(df[features].values, tf.float32),
                        tf.cast(df['code'].values, tf.int32)
                    )
                )
            ).shuffle(100)

    predictions = model.predict(test_ds)
    
    return predictions
```

```python
def PrintHasil(df, model_path):
    predictions = Pengujian(df, model_path)
    code = list(df['code'].values)
    label = list(df['label citra'].values)
    
    for i in range(len(predictions)):
        if code[i] == np.argmax(predictions[i]):
            print("citra",df.index[i],"berhasil diidentifikasi")
            print("Hasil prediksi: ", label[i])
            print()
        else:
            print("citra",df.index[i]," gagal diidentifikasi sebagai: ", label[i])
            print("--->> Hasil prediksi: ", df_main.loc[df_main['code']==np.argmax(predictions[i]), 'label citra'].iloc[0])
            print()
```
